```python
data_path =  '/content/drive/MyDrive/ASSIGNMENT-3/Obesity Dataset.csv'
df = pd.read_csv(data_path)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']= le.fit_transform(df['Gender'])
df['family_history_with_overweight']= le.fit_transform(df['family_history_with_overweight'])
df['FAVC']= le.fit_transform(df['FAVC'])
df['CAEC']= le.fit_transform(df['CAEC'])
df['SMOKE']= le.fit_transform(df['SMOKE'])
df['SCC']= le.fit_transform(df['SCC'])
df['CALC']= le.fit_transform(df['CALC'])
df['MTRANS']= le.fit_transform(df['MTRANS'])
df['NObeyesdad']= le.fit_transform(df['NObeyesdad'])
df
```

```python
X=df.drop(['NObeyesdad'], axis=1)
X
Y=df['NObeyesdad']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=41)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree

model1=LogisticRegression(random_state=1)
model2=GaussianNB()
model3=tree.DecisionTreeClassifier(random_state=1)
```

```python
from sklearn.ensemble import VotingClassifier
e_model= VotingClassifier(estimators=[('LR',model1),('NB',model2),('DT',model3)], voting='soft')
```

```python
[c.fit(x_train,y_train) for c in (model1,model2,model3,e_model)]
```

```python
from sklearn import metrics
y_pred= model1.predict(x_test)
print('MODEL 1 ACC', metrics.accuracy_score(y_test,y_pred))

y_pred= model2.predict(x_test)
print('MODEL 2 ACC', metrics.accuracy_score(y_test,y_pred))

y_pred= model3.predict(x_test)
print('MODEL 3 ACC', metrics.accuracy_score(y_test,y_pred))

y_pred= e_model.predict(x_test)
print('e_model ACC', metrics.accuracy_score(y_test,y_pred))
```

```python
estimators=[('LR',model1),('KNN',model2),('DT',model3)]
```

```python
from sklearn.ensemble import RandomForestClassifier
final_model=RandomForestClassifier()
final_estimator=final_model
```

```python
from sklearn.ensemble import StackingClassifier
stacked_clf= StackingClassifier(estimators=estimators, final_estimator=final_estimator)
stacked_clf.fit(x_train,y_train)
```

**Strength of Ensemble ** :- Ensemble methods often outperform individual models by combining the predictions of multiple base models. This can lead to higher accuracy and better generalization on unseen data.By combining multiple models, ensemble methods can mitigate overfitting. Ensemble methods are flexible and can incorporate a wide range of base models, including classifiers, regressors, and even more complex models like neural networks. This versatility allows ensemble techniques to be applied to various types of problems across different domains.

**Weakness of Ensemble** :- Stacking involves training multiple base models and a meta-model, which can increase computational overhead, especially when dealing with large datasets or complex models. Ensemble techniques often involve tuning multiple hyperparameters for each base model and the meta-model. Finding the optimal hyperparameters can be challenging and may require extensive experimentation, increasing the overall complexity of the model selection process.
